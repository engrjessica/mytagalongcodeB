```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
new_df = pd.get_dummies(df)
new_df
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
df['date'].unique(), df['Appliances'].unique(), df['lights'].unique()
```

```python
x = new_df.drop('Appliances', axis = 1)
y = new_df['Appliances']
```

```python
x_train, x_test, y_train, y_test, = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
models = {
          "DecisionTree Regressor": DecisionTreeRegressor(),
          "Random Forest": RandomForestRegressor(),
          "Ridge": Ridge(alpha=0.5),
          "Lasso": Lasso(alpha=0.5)
         }

def fit_and_score(models, x_train, x_test, y_train, y_test):
    
    np.random.seed(42)
    
    model_scores = {}
    
    for name, model in models.items():
        
        model.fit(x_train, y_train)
        
        model_scores[name] = model.score(x_test, y_test)
    return model_scores
```

```python
scores = fit_and_score(models=models,
                             x_train=x_train,
                             x_test=x_test,
                             y_train=y_train,
                             y_test=y_test)
scores
```

```python
model_compare = pd.DataFrame(scores, index=['R2_score'])
model_compare.T.plot.bar();
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()

x_train_scaled = Scaler.fit_transform(x_train)
x_test_scaled = Scaler.transform(x_test)
```

```python
model2 = RandomForestRegressor()
model2.fit(x_train_scaled, y_train)
```

```python
y_preds = model2.predict(x_test_scaled)
y_preds
```

```python
mse = mean_squared_error(y_test, y_preds)
mse
```

```python
np.sqrt(mse)
```

```python
mae = mean_absolute_error(y_test, y_preds)
mae
```

```python
df2 = pd.DataFrame(data={"actual values": y_test,
                        "predictions": y_preds})

df2
```

```python
def plot_importances(columns, importances, n=10):
    df3 = (pd.DataFrame({'features': columns,
                        'feature_importances': importances})
                        .sort_values('feature_importances', ascending=False)
                        .reset_index(drop=False))
    
    fig, ax = plt.subplots(figsize=(10,5))
    ax = sns.barplot(x = 'feature_importances', y = 'features', data = df3[:n], orient='h')
    plt.ylabel('features')
    plt.xlabel('feature_importances')
```

```python
plot_importances(x_train.columns, model2.feature_importances_)
```
